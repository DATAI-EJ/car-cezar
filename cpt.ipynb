## Pegando informações do CPT

```python
import unidecode
import pandas as pd
import geopandas as gpd


def listar_mun(column, sep):  
    muns = set()
    for i in column.__iter__():
        if i.find(sep) != -1:
            i = i.split(sep)
            for j in i:
                muns.add(j)
        else:
            muns.add(i)
    return muns

def remover_acentos(texto):
    return unidecode.unidecode(texto)
```

```python
cnuc = gpd.read_file("areas-PA/áreas-selecionadas/cnuc/cnuc.shp")
cnuc
```

Pegando os municípios

```python
muns = listar_mun(cnuc["municipio"], ' , ')
muns
```

```python
muns.remove("trairao")
```

```python
def count_muns(col):
    muns_count = {"altamira":0, "itaituba":0, "jacareacanga":0, "novo progresso":0, "sao felix do xingu":0}
    for mun in muns_count.keys():
        for i in col:
            if i.find(mun) != -1:
                muns_count[mun] += 1
    return muns_count
```

### Ameaçados

```python
ameacados = pd.read_excel("CPT/CPT.xlsx", sheet_name="Ameaçados")
ameacados_filter = ameacados[ameacados["UF"] == "PA"]
ameacados_filter
```

### Áreas em conflito

```python
areas_conflito = pd.read_excel("CPT/CPT.xlsx", sheet_name="Áreas em Conflito")
areas_conflito["mun"] = areas_conflito["Municípios"].apply(remover_acentos)
areas_conflito["mun"] = areas_conflito["mun"].str.lower()

areas_conflito_filter = areas_conflito[areas_conflito["mun"].isin(muns)]
```

```python
areas_conflito_filter = pd.DataFrame(columns=areas_conflito.columns)
for mun in muns:
    areas_conflito_filter = pd.concat([areas_conflito_filter, areas_conflito[areas_conflito["mun"].str.contains(mun)]])

areas_conflito_filter
```

```python
conflitos_count = count_muns(areas_conflito_filter["mun"])
conflitos_count
```

```python
areas_conflito_filter.groupby("mun").sum()
```

```python
assasinatos_count = count_muns(assassinatos_filter["mun"])
assasinatos_count
```

### Conflitos por Terra

```python
def separar_estado(mun):
    return mun.split("(")[0]

conflitos = pd.read_excel("CPT/CPT.xlsx", sheet_name="Conflitos por Terra")
conflitos["mun"] = conflitos["Município"].apply(separar_estado)
conflitos["mun"] = conflitos["mun"].str.lower()
conflitos["mun"] = conflitos["mun"].apply(remover_acentos)

conflitos_filter = conflitos[conflitos["mun"].isin(muns)]
```

```python
terra_count = count_muns(conflitos_filter["mun"])
terra_count
```
